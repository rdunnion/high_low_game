```python
import random

NUM_ROUNDS = 5
score = 0

print("Welcome to the High-Low Game! --------------------------------")
for i in range(1,NUM_ROUNDS+1):
    ryan_num = random.randint(1, 100)
    pc_num = random.randint(1,100)

    print("Round", i)
    print("Your number is:", ryan_num)
    guess = input("Do you think your number is higher or lower than the computer's?:").lower().strip()

    while guess != "lower" and guess != "higher":
        guess = input("Please choose either higher or lower:").lower().strip()
    
    if guess == "higher" and ryan_num > pc_num:
        print("You were right! The computer's number was", pc_num)
        score = score + 1

    elif guess == "lower" and ryan_num < pc_num:
        print("You were right! The computer's number was", pc_num)
        score = score + 1

    else: 
        print("You were incorrect. The computer's number was", pc_num)
        
    print("Your score is:", score)

if score == NUM_ROUNDS:
    print("... Your score is now:", score)
    print("Wow! You played perfectly!")

elif score >= int(NUM_ROUNDS/2):
    print("... Your score is now:", score)
    print("Good job, you played really well!")

else:
    print("... Your score is now:", score)
    print("Better luck next time!")

print("Thanks for playing!")
```
